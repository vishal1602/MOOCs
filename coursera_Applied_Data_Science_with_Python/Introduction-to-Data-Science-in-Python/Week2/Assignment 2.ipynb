```python
df = pd.read_csv(data_folder + 'olympics.csv', skiprows=1, index_col=0)

# fix column names
for col in df.columns:
    if col[:2] == '01':
        df = df.rename(columns={col: 'Gold'+col[4:]})
    elif col[:2] == '02':
        df = df.rename(columns={col: 'Silver'+col[4:]})
    elif col[:2] == '03':
        df = df.rename(columns={col: 'Bronze'+col[4:]})
    else:
        df = df.rename(columns={col: col.replace('№','#')})

# fix index
name_ids = df.index.str.split('(')
country_name = []
for name in name_ids:
    country_name.append(name[0])
df.index=country_name
```

### Question 5
Which state has the most counties in it? (hint: consider the sumlevel key carefully! You'll need this for future questions too...)

```python
county = census_df.groupby('STNAME').agg({'COUNTY':sum})
county.idxmax().values[0]
```

### Question 6
Only looking at the three most populous counties for each state, what are the three most populous states (in order of highest population to lowest population)? Use CENSUS2010POP.

This function should return a list of string values.

```python
census_df[census_df.SUMLEV == 50].groupby('STNAME').CENSUS2010POP.agg({'CENSUS2010POP':
                                                                       'sum'}).sort_values('CENSUS2010POP', 
                                                                                   ascending=
                                                                                   False)[:3].index.values.tolist()
```
